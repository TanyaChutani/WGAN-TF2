```
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

def test_show_image(generator_model):
  latent_dim = 128

  test_latent_dim = tf.random.normal(shape=(10,latent_dim))

  plt.figure(figsize=(15,15))
  for i in range(10):
    plt.subplot(5,2,i+1)
    predicted_image = generator_model(test_latent_dim,training=False)
    plt.imshow(tf.squeeze(predicted_image[i]),cmap='gray')
    plt.subplots_adjust(wspace=1, hspace=1)
  plt.show()

generator_model = tf.keras.models.load_model('wgan_checkpoint/my_model')

test_show_image(generator_model)
```
